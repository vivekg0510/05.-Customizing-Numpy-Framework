```python
# Import numpy as np
import numpy as np
import pandas as pd

baseball = pd.read_csv("../Numpy/baseball.csv")
np_height = np.array(baseball.Height)
np_baseball = np.array(baseball)
```

```python
# For loop over np_height
for x in np_height:
    print(str(x) + " inches")
```

```python
# For loop over np_baseball
for x in np.nditer(np_baseball):
    print(x)
```
