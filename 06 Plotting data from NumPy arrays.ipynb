```python
# Import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
# Create array using np.linspace: x
x = np.linspace(0, 5, 100)
```

```python
# Create array using np.cos: y
y = np.cos(x)
```

```python
# Add circles at x and y
p = plt.pie(x, y)

# Specify the name of the output file and show the result
plt.show(p)
```
